```python
# Renaming columns

customers_raw.rename(columns = {'First Name' : 'first_name'}, inplace = True)
customers_raw.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
customers_raw.rename(columns = {'Gender' : 'gender'}, inplace = True)
customers_raw.rename(columns = {'STATE' : 'state'}, inplace = True)
customers_raw.rename(columns = {'Age' : 'age'}, inplace = True)
customers_raw.rename(columns = {'n_dependants' : 'num_dependants'}, inplace = True)
customers_raw.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```

```python
# Check changes
customers_raw.head()
```

```python
# Checking missing values

missing_values = customers_raw.isnull().sum()
missing_values
```

```python
# Using iloc to get the values where 'first_name' is null

df_null_first_name = customers_raw.loc[customers_raw['first_name'].isnull()]

# Display the 'user_id', 'last_name', and 'first_name' columns for the filtered rows

result_table = df_null_first_name[['user_id', 'last_name', 'first_name']]

print(result_table)
```

```python
# The number of missing values by customer name is 11259.
# If user_id is unique, even with the missing information, It is possible to work with the data frame without taking any additional actions.
```

```python
# finding duplicate rows
customers_dup = customers_raw.duplicated()
```

```python
# summing up the total number of duplicates
num_duplicates = customers_dup.sum()
```

```python
# output results: No duplicates
num_duplicates
```

```python
# Looking for mixed-type data

for col in customers_raw.columns:
  mix_data = (customers_raw[[col]].map(type) != customers_raw[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers_raw[mix_data]) > 0:
    print (col)
```

```python
# Changing the data type in the column 'first_name' to string

customers_raw['first_name'].astype('str')
```

```python
# Checking the data type in the column 'first name'

customers_raw['first_name'].dtype
```

```python
# For ethical reasons, clients' first and last names will not be included in the data analysis. 
# They don't need to be removed from the table. However, it is noted that the 'first_name' column contains a large number of missing values. 
# Therefore, to ensure data cleanliness, it can be removed for this analysis.
```

```python
# Removing the 'first_name' column

customers_new = customers_raw.drop('first_name', axis=1)

# Checking the changes

customers_new.head()
```

## Combine your customer data with the rest of your prepared Instacart data. (Hint: Make sure the key columns are the same data type!)

```python
# importing dataframe
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_grouped.pkl'))
```

```python
# Checking the data

ords_prods_merge.head()
```

```python
# Dropping the column '_merge'

df_ords_prods_merge = ords_prods_merge.drop(columns = ['_merge'])

#Checking the changes

df_ords_prods_merge.head()
```

```python
# Checking the shape

customers_new.shape
```

```python
# Checking if there is the same amount of user_id in both datasets

df_ords_prods_merge['user_id'].value_counts()
```

```python
# Merging customer with ords_prods_merge

df_merged_all = df_ords_prods_merge.merge(customers_new, on = 'user_id', indicator = True)
df_merged_all.head()
```

```python
# Checking the shape

df_merged_all.shape
```

```python
# Checking if there is a full match

df_merged_all_outer = df_ords_prods_merge.merge(customers_new, on = 'user_id', how = 'outer', indicator = True)
df_merged_all_outer['_merge'].value_counts() 
```
